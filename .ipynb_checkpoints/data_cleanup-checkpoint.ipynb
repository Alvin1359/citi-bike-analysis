```python
import pandas as pd
import glob
import os
```

```python
path = os.getcwd()
filenames = glob.glob(os.path.join(path + "/resources", "*.csv"))
  
dfs = []

# loop over the list of csv files
for filename in filenames:

    dfs.append(pd.read_csv(filename, index_col=None, header=0))

# Concatenate all data into one DataFrame

data_2020 = pd.concat(dfs, axis=0, ignore_index=True)
data_2020
```

```python
# Checking for NaN values
data_2020.isnull().sum().sum()
```

```python
# Changing encoders for gender (0 = Unknown, 1 = male, 2 = female)
data_2020['gender']= data_2020['gender'].map({0: "Unknown", 1: "Male", 2: "Female"})
```

```python
# Formatting time columns to pandas datetime format
data_2020["starttime"]= pd.to_datetime(data_2020["starttime"])
data_2020["stoptime"]= pd.to_datetime(data_2020["stoptime"])
```

```python
# Checking for spelling mistakes in station names
data_2020["start station name"].value_counts()
```

```python
data_2020.to_csv("resources/2020-citibike-tripdata.csv", index=False)
```
